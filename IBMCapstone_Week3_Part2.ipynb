### Part2: Determining the Borough Coordinates using Toronto Postcodes
#### Since geocoder was very problematic to work with I used instead the pgeocode package in order to determine the coordinates of each postcode

```python
import pandas as pd
import numpy as np
import pgeocode
pd.options.display.max_rows = 999
pd.set_option('max_colwidth', 140)
```

Inserting the data frame from a csv file that was created previously

```python
df_toronto=pd.read_csv('PostCodes_Toronto_noIndex.csv')
df_toronto
```

Creating the two extra columns with names Longitude and Latitude

```python
df_toronto=df_toronto.assign(Latitude=0)
df_toronto=df_toronto.assign(Longitude=0)
df_toronto
```

Using the pgeocode package in order to extract the longitude and latitude values for each  postcode

```python
canada = pgeocode.Nominatim('ca')

for index,postcode in enumerate(df_toronto['Postcode']):
    postcode_query=0
    postcode_query=canada.query_postal_code(postcode)
    
    lat=round(postcode_query.iloc[9],4)
    long=round(postcode_query.iloc[10],4)
    
    df_toronto.at[[index],'Latitude']=lat
    df_toronto.at[[index],'Longitude']=long
    
df_toronto
```

Remark: The postcode M7R returned no coordinates. This must be due to the package i used since google maps are returning the proper facility. The coordinates found on line are 43.6370 and -79.6158 and we will import them manually in the dataframe.

```python
df_toronto.at[[65],'Latitude']=43.6370
df_toronto.at[[65],'Longitude']=-79.6158
```
